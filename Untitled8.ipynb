```python
# Tarih sütununu datetime formatına çevirme
df['tarih'] = pd.to_datetime(df['tarih'], format='%d.%m.%Y')

# Tarih bileşenlerine ayırma
df['year'] = df['tarih'].dt.year
df['month'] = df['tarih'].dt.month
df['day'] = df['tarih'].dt.day
```

```python
# Kategorik değişkenleri sayısal hale getirme
le_hakim_rüzgar_yönü_hız = LabelEncoder()
le_max_rüzgar_yönü = LabelEncoder()
le_hakim_rüzgar_yönü_esneme = LabelEncoder()
le_sel_durumu = LabelEncoder()
df['Aylık Hakim Rüzgar Yönü- Rüzgar Hızı '] = le_hakim_rüzgar_yönü_hız.fit_transform(df['Aylık Hakim Rüzgar Yönü- Rüzgar Hızı '])
df['Aylık Maksimum Rüzgar Yönü'] = le_max_rüzgar_yönü.fit_transform(df['Aylık Maksimum Rüzgar Yönü'])
df['Aylık Hakim Rüzgar Yönü - Esneme Sayısı'] = le_hakim_rüzgar_yönü_esneme.fit_transform(df['Aylık Hakim Rüzgar Yönü - Esneme Sayısı'])
df['SINIF'] = le_sel_durumu.fit_transform(df['SINIF'])
```

```python
#öncelikli olarak histogram grafikleini çizdireceğimiz fonksiyonu yazalım.
def plotHistogram(degisken):
    """
        Girdi: Değisken / sütun ismi
        Çıktı: İlgili değişkenin histogramı
    """
    plt.figure()
    plt.hist(df[degisken], bins=75, color = "blue")
    plt.xlabel(degisken)
    plt.ylabel("Frekans")
    plt.title("Veri Sıklığı - {}".format(degisken))
    plt.show()
```

```python
#SVM
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
# SVM Confusion Matrix Görselleştirme
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', xticklabels=le_sel_durumu.classes_, yticklabels=le_sel_durumu.classes_)
plt.title('SVM Confusion Matrix')
plt.xlabel('Tahmin Edilen Sınıf')
plt.ylabel('Gerçek Sınıf')
plt.show()

# SVM Accuracy Score Görselleştirme
plt.figure(figsize=(10, 6))
plt.bar(range(1, 11), accuracies, color='green', alpha=0.7)
plt.title('SVM Model Cross-Validation Accuracies')
plt.xlabel('Fold Number')
plt.ylabel('Accuracy')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# SVM doğruluk oranı ve standart sapması
accuracy = 67.33
std_dev = 16.26

# Grafik çizdirme
labels = ['Accuracy']
x = np.arange(len(labels))  # label konumları
width = 0.35  # çubukların genişliği

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, accuracy, width, label='Accuracy', color='b')
rects2 = ax.bar(x + width/2, std_dev, width, label='Standard Deviation', color='r')

# Eksenleri isimlendirme, başlık ve özel legend ekleme
ax.set_ylabel('Values')
ax.set_title('Accuracy and Standard Deviation of SVM Model')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Çubukların üzerine değer yazdırma fonksiyonu
def autolabel(rects):
    """Çubukların üzerine etiket ekleyen yardımcı fonksiyon."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Çubuklara değer yazdır
autolabel(rects1)
autolabel(rects2)

# Grafiği göster
plt.show()
```
